# Posição do mouse 

```python
time.sleep(2) #usado pra esperar um sistema abrir

posicao_mouse = pyautogui.position()

print(posicao_mouse) #importante lembrar que isso varia de um tamanho de monitor para outro e, claro, pode afetar o funcionamento correto
```

# Click


```python
pyautogui.click(x=posicao_mouse[0], y=posicao_mouse[1], 
                clicks=1, interval=0, button='left') # Parametros de ajuste. 

```

# Movendo o mouse de posição na tela

```python
pyautogui.moveTo(x=posicao_mouse[0], y=posicao_mouse[1], duration=1)  # permite ver o movimento do mouse.
```

# Ponto importante:

O sistema tem que ter sempre a mesma cara. Você só vai pegar as coordenadas uma vez.

"Mas e páginas que tem que scrollar?"

Sim, cada scroll, a coordenada muda, mas se a página precisa scrollar você tem duas opções:

* Ta usando a biblitoca errada
* Vai scrollar com o pyautogui também. Você pode contar quantos scrolls pra baixo ou pra cima o programa da.

```python
import subprocess
import pyautogui
import time

# Substitua o caminho abaixo pelo caminho correto do executável do Brave em seu sistema
brave_path = r"C:\Program Files\BraveSoftware\Brave-Browser\Application\brave.exe"

# Abrir o navegador Brave usando subprocess
subprocess.Popen([brave_path])

# Aguardar um momento para o navegador abrir completamente
time.sleep(5)

# Mova o mouse para a barra de endereço
pyautogui.moveTo(x=posicao_mouse[0] + 200, y=posicao_mouse[1], duration=1)

# Clique para focar na barra de endereço
pyautogui.click()

# Digite o endereço desejado
pyautogui.write("www.investing.com")

# Pressione Enter para acessar o site
pyautogui.press("enter")
```
